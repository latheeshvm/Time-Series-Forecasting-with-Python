```python
import numpy as np
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import month_plot, quarter_plot, plot_acf
import matplotlib.pyplot as plt
```

```python
#load the data
# YYYY-MM-DD
df = pd.read_csv('nyc_data.csv', index_col=0, parse_dates = True)
df.head()
```

```python
# Rename heading
df = df.rename(columns = {'Demand': 'y'})
df.head(0)
```

```python
# Summary statics
df.describe().transpose()
```

```python
# Sesonal Decomposition

decomposition = seasonal_decompose(x = df.y, model = "multiplicative", period =365)
decomposition.plot();
```

```python
decomposition.resid.plot();
```

```python
# seasonal graphs
month_plot(df.y.resample(rule = "MS").mean());
quarter_plot(df.y.resample(rule = "Q").mean());
```

```python
# advanced visulisation

# create a version B of the datafrom

#df_plot = df.loc['2015-01-01':'2015-12-31']
df_plot = df

# Define the colors
temp_color = 'red'
y_color = 'Black'

#plot 
fig, ax1 = plt.subplots(figsize = (16,6))
ax2  = ax1.twinx()


#data and lines
ax1.plot(df_plot.index, df_plot.y, color = y_color, lw = 1)
ax2.plot(df_plot.index, df_plot.Temperature, color = temp_color, lw = 1)


#title and dates
fig.suptitle("The relationship between temprature and demand", fontsize = 20)
fig.autofmt_xdate()

#first axis
ax1.set_label("Date")
ax1.tick_params(axis="y",labelcolor = y_color)
ax1.set_ylabel("Demand", color = y_color)


#second axis
ax2.set_label("Date")
ax2.tick_params(axis="y",labelcolor = temp_color)
ax2.set_ylabel("Temprature", color = temp_color)

events = df_plot[(df_plot.Easter == 1 ) | (df_plot.Christmas == 1)].index
ax1.vlines(x = events, ymin = 500, ymax = 800, color = 'k', ls = "--")
```
